# 4. Remove Values

```python
df_clean = df
df_clean = df_clean.drop('Cabin',1) # 1 ในที่นี้คือดรอป Column
df_clean = df_clean.drop('Ticket Number',1)
df_clean.columns
```

```python
df_clean = df_clean.dropna(1,'any')
# 1 = ลบคอลลัมใดๆ('any')ที่มีค่าnull
```

```python
df_clean = df_clean.dropna(0,'any')
# 0 = ลบrowใดๆ('any')ที่มีค่าnull
```

```python
df_clean.isnull().any()
```

# 5. Replace values

```python

df_clean['Sex'].unique()
```

```python
df_clean['Sex'] = df_clean['Sex'].replace('Female','F')
df_clean['Sex'] = df_clean['Sex'].replace('Male','M')
df_clean
```

# 6.Replace string values to unique integers

```python
p_class, levels = pd.factorize(df_clean['Passenger Class'])
#pd.factorize คือ เปลี่ยนข้อมูล string -> ตัวเลข
```

```python
p_class
```

```python
levels
```

```python
df_clean['Passenger Class'] = p_class
df_clean
```

```python
df_clean.dtypes
```

```python
df_clean['Sex'] = pd.factorize(df_clean['Sex'])[0] 
df_clean['Survived'] = pd.factorize(df_clean['Survived'])[0]
df_clean.dtypes
# [0] หลังสุดคือนำค่าที่แปลงแล้วเฉพาะเลขมาเท่ากับ df_clean['Sex'] เพราะโดยปกติ
# เมื่อใช้คำสั่ง pd.factorize จะได้ข้อมูลม 2 array คือ เลข,string
```

# 7. Set Index

```python
df_clean = df_clean.set_index('Name')
df_clean
```

# 8.Cross Tab Plot

```python
pd.crosstab(df_clean['Passenger Class'],df_clean['Survived']).plot(kind='bar')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plenoi/Clinic/master/ultima_all_clean.csv',sep=',',header = 0)
df
```

```python
df= df.set_index('hn')
df
```

# 2.) Remove missing values

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

# 3.) Remove these unnecessary columns hiv, size, utmet, vgmet, surgery, pchemo

```python
df_clean = df
df_clean = df_clean.drop('hiv',1)
df_clean = df_clean.drop('size',1)
df_clean = df_clean.drop('utmet',1)
df_clean = df_clean.drop('vgmet',1)
df_clean = df_clean.drop('surgery',1)
df_clean = df_clean.drop('pchemo',1)
df_clean
```

```python

df_clean = df_clean.dropna(0,'any')

```

```python
df_clean.isnull().any()
```

```python
df_clean.dtypes
```

```python
df_clean = df_clean.astype({"age" : str})
df_clean.dtypes
```

# 4.) Change age into 2 values 'Young' (age <= 20) and 'Old' (age > 20)

```python
import numpy as np
df_clean['age'] = np.where(df_clean['age'] > '20','Old','Young')
df_clean['age']
```

```python
df_clean['pmmet'].unique()
```

# 5.) Separate data into groups based on pmmet values

```python
group1 = df_clean[df_clean['pmmet'] == 0.0]
group2 = df_clean[df_clean['pmmet'] == 1.0]
group1
group2
```

# 6.) for each group, randomly choose 10 data

```python
group1 = group1.sample(n=10, random_state=1)
group1
```

```python
group2 = group2.sample(n=10, random_state=1)
group2
```
