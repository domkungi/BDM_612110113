# 2. Sub Plot

```python
plt.figure(1, figsize=(15,3)) #กำหนด size โดย 1 ตัวแรกคือชื่อเฉยๆ 

plt.subplot(1,3,1) #1,3 = ใน 1 แถว มี 3 รูป และ 1 ตัวสุดท้าย คือ ไป plot ในรูปที่ 1
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x$') #ใช้คำสั่งในการเขียนตัวอักษรพิเศษ
plt.plot(y1, color='tan')

plt.plot(y1, color='tan')        
plt.subplot(1,3,2)               
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^2$')      
plt.plot(y2, color='darkorange')

plt.subplot(1,3,3)              
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'$f(x) = x^3$')
plt.plot(y3, color='green')      

plt.show()
```

# 3. Graph Type

```python
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(15,6))

#bar chart 
accounts = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
steem = [1100, 500, 1200, 2300]

plt.subplot(1,2,1)
plt.title('Bar chart')
plt.ylabel('Steem')
plt.bar(accounts, steem)

#pie chart
labels = ['fabiyamada', 'jedigeiss', 'scipio', 'stoodkev']
perc = [steem[0]/sum(steem),
        steem[1]/sum(steem),
        steem[2]/sum(steem),
        steem[3]/sum(steem)]

explode = (0, 0, 0.1, 0)

plt.subplot(1,2,2)
plt.title('Pie chart')
plt.axis('equal')
plt.pie(perc, labels=labels, autopct='%1.2f%%',
       startangle=90, explode=explode)

plt.show()
```

# 4. Load Data & Basic Data Preprocessing

```python
import numpy as np
titanics = np.genfromtxt('/kaggle/input/titanic.txt',delimiter='\t',missing_values = None,dtype=str)
```

```python
titanics[0,:]
```

```python
titanics[:,3]
```

```python
titanics = np.delete(titanics,[1,6,7,8,9,10],1)
```

```python
titanics[0,:]
```

```python
rows, cols = np.where(titanics == None)
rows
```

```python
rows, cols = np.where(titanics == '')
np.unique(rows)
```

```python
np.unique(cols)
```

```python
titanics = np.delete(titanics,2,1)
titanics[0,:]
```

```python
passenger_class = np.unique(titanics[1:,0])
passenger_class
```

```python
passenger_class_value = [sum(titanics[:,0]==passenger_class[0]),
                         sum(titanics[:,0]==passenger_class[1]),
                         sum(titanics[:,0]==passenger_class[2]) ]
passenger_class_value
```

```python
plt.bar(passenger_class, passenger_class_value)
```
